7. Crie um algoritmo que receba o valor do salário do funcionário por hora e quantas horas ele trabalha por semana, levando em consideração que todo mês tem 4 semanas. Serão descontados 8% do INSS no salário mensal do funcionário. Com isso, o algoritmo deve apresentar:

Salário bruto por semana;
Salário bruto mensal;
Valor do INSS em reais;
Salário líquido mensal (líquido = bruto – impostos).

```python
#ex7
salario_h =  float(input("Digite o valor pro hora de trabalho: "))
horas_trabalho = int(input("Digite a quantidade de horas trabalho semanais: "))
salario_semanal = salario_h * horas_trabalho
salario_mensal = salario_semanal * 4
porcentagem = 8 / 100
inss = salario_mensal * porcentagem
salario_liquido = salario_mensal - inss

print(f"Salário semanal: R$ {salario_semanal:.2f}")
print(f"Salário mensal: R$ {salario_mensal:.2f}")
print(f"O valor de desconto do INSS é: R$ {salario_semanal:.2f}")
print(f"Salário líquido R$ {salario_liquido:.2f}")
```

8. O usuário deve digitar um número inteiro com 6 dígitos. Calcule a soma desses 6 dígitos. Ex: 123456 = 21.0
